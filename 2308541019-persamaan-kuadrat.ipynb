```python
import math

def solve_quadratic(a, b, c):
    # Menghitung diskriminan
    discriminant = b**2 - 4*a*c

    # Memeriksa nilai diskriminan
    if discriminant > 0:
        # Dua akar real berbeda
        x1 = (-b + math.sqrt(discriminant)) / (2*a)
        x2 = (-b - math.sqrt(discriminant)) / (2*a)
        return x1, x2
    elif discriminant == 0:
        # Akar ganda
        x = -b / (2*a)
        return x
    else:
        # Akar imajiner
        real_part = -b / (2*a)
        imaginary_part = math.sqrt(abs(discriminant)) / (2*a)
        return complex(real_part, imaginary_part), complex(real_part, -imaginary_part)

# Input koefisien a, b, dan c dari pengguna
a = float(input("Masukkan koefisien a: "))
b = float(input("Masukkan koefisien b: "))
c = float(input("Masukkan koefisien c: "))

# Memanggil fungsi solve_quadratic
result = solve_quadratic(a, b, c)

# Output hasil
if isinstance(result, tuple):
    print("Akar persamaan kuadrat:")
    for root in result:
        print("x =", root)
elif isinstance(result, complex):
    print("Akar persamaan kuadrat:")
    print("x1 =", result)
    print("x2 =", result.conjugate())
else:
    print("Akar persamaan kuadrat:")
    print("x =", result)
    
# Membuat plot
plt.figure()
plt.plot(x, y, label=f'y = {a}x^2 + {b}x + {c}')
plt.title('Grafik Persamaan Kuadrat')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_function(x, a, b, c):
    return a * x**2 + b * x + c

# Input koefisien a, b, dan c dari pengguna
a = float(input("Masukkan koefisien a: "))
b = float(input("Masukkan koefisien b: "))
c = float(input("Masukkan koefisien c: "))

# Menghitung nilai x dari -10 hingga 10
x = np.linspace(-10, 10, 400)

# Menghitung nilai y menggunakan fungsi kuadratik
y = quadratic_function(x, a, b, c)

# Membuat plot
plt.figure()
plt.plot(x, y, label=f'y = {a}x^2 + {b}x + {c}')
plt.title('Grafik Persamaan Kuadrat')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```
